```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head(3)
```

```python
df_test.head(3)
```

```python
df_train.drop('Name', axis=1, inplace=True)
df_test.drop('Name', axis=1, inplace=True)
```

```python
df_train.drop('PassengerId', axis=1, inplace=True)
df_test.drop('PassengerId', axis=1, inplace=True)
```

```python
df_train.drop('Ticket', axis=1, inplace=True)
df_test.drop('Ticket', axis=1, inplace=True)
```

```python
df_train.drop('Cabin', axis=1, inplace=True)
df_test.drop('Cabin', axis=1, inplace=True)
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
df_train['Age'].fillna(df_train['Age'].value_counts().index[0], inplace=True)
```

```python
df_train['Embarked'].fillna(df_train['Embarked'].value_counts().index[0], inplace=True)
```

```python
df_test['Age'].fillna(df_test['Age'].value_counts().index[0], inplace=True)
```

```python
df_test['Fare'].fillna(df_test['Fare'].value_counts().index[0], inplace=True)
```

```python
for column in df_train.columns:
    if df_train[column].dtype == object:
        label = LabelEncoder()
        df_train[column] = label.fit_transform(df_train[column])
        df_test[column] = label.transform(df_test[column])
```

```python
model = LogisticRegression()
```

```python
scale = StandardScaler()
```

```python
pipe = Pipeline([('scale', scale), ('model', model)])
```

```python
params = [
    {
        'model__penalty': ['l1'],
        'model__solver': ['liblinear', 'saga'],
        'model__max_iter': [100, 500, 1000]
    },
    {
        'model__penalty': ['l2'],
        'model__solver': ['liblinear', 'lbfgs', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'model__max_iter': [100, 500, 1000]
    },
    {
        'model__penalty': ['elasticnet'],
        'model__solver': ['saga'],
        'model__max_iter': [100, 500, 1000],
        'model__l1_ratio': [0.15, 0.5, 0.85]
    },
    {
        'model__penalty': ['none'],
        'model__solver': ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'model__max_iter': [100, 500, 1000]
    }
]
```

```python
final_model = GridSearchCV(pipe, param_grid=params, verbose=2, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
X_train = df_train.drop('Survived', axis=1)
X_test = df_test.copy()
y = df_train['Survived']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.1, random_state=42)
```

```python
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_val)
```

```python
print(classification_report(y_pred, y_val))
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_pred, y_val)).plot()
```
